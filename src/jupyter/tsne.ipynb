```python
import pandas as pd
import chemplot as cp
import matplotlib.pyplot as plt


df = pd.read_csv('../../Data/SAC/chemdiv_bb.csv')
plotter = cp.Plotter.from_smiles(df["SMILES"].sample(5000))

plotter.tsne(perplexity=30)

x = plotter.visualize_plot()
x.set_facecolor('white')
x.savefig('../../figures/input_tsne.svg', format='svg', dpi=1200)
```
